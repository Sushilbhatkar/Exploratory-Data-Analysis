```python
def encode(zomato):
    for column in zomato.columns[~zomato.columns.isin(['rate','votes','cost'])]:
        zomato[column]=zomato[column].factorize()[0]
    return zomato
```

```python
zomato2_encoded=encode(zomato2)
```

```python
plt.rcParams['figure.figsize']=(10,10)
y=pd.crosstab(zomato3['rate'],zomato3['book_table'])
y.div(y.sum(axis=1).astype(float),axis=0).plot(kind='bar',stacked=True,color=['red','blue'])
plt.title("Table Booking Vs Rating of Restaurants",fontweight=30,fontsize=15)
plt.legend(loc="upper right")
plt.show()
```

```python
#Can a Table Booking affect a Resturants Cost?
plt.rcParams['figure.figsize']=(10,10)
z=pd.crosstab(zomato3['cost'],zomato3['book_table'])
z.div(z.sum(axis=1).astype(float),axis=0).plot(kind='bar',stacked=True,color=['red','blue'])
plt.title("Table Booking Vs Cost of the Restaurants",fontweight=30,fontsize=15)
plt.legend(loc="upper right")
plt.show()
```
