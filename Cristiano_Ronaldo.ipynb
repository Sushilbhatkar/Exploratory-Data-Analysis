## Handling Missing Values

```python
#Filling NaN values in Column "remaining_sec"with MEAN

df['power_of_shot'].fillna(value=df['power_of_shot'].mean(),inplace=True)
df.isnull().sum()
```

```python
#Filling NaN values in Column "type_of_combined_shot" with MODE
mode_com = df.type_of_combined_shot.value_counts().keys()[0]
print('moded is: ',mode_com)
df.type_of_combined_shot.fillna(value=mode_com,inplace=True)
df.isnull().sum()  #number of missing values for type_of _combined_shot column should be zero
```

```python
#filling NaN values in Column "remaining_sec" with MEDIAN

df.remaining_sec.fillna(value=df.remaining_sec.median(),inplace=True)
df.isnull().sum()
```

```python
#shot_id_no.
df.shot_id_number = pd.Series(np.arange(1,df.shot_id_number.shape[0]+1))
df.isnull().sum()
```

```python
#Filling NaN values in columns 'Location_x' and Location_y with 0
df['location_x'].fillna(value=0, inplace = True)
df['location_y'].fillna(value= 0, inplace = True)
df.isnull().sum()
```

```python
#Using Forward Filling method in appropriate Columns
print('Null values in column home/away before forward fill =',df['home/away'].isnull().sum())

col=['home/away','lat/lng','team_name','match_id','match_event_id','team_id','remaining_min','knockout_match','game_season']

df.loc[:,col]=df.loc[:,col].ffill()

print('Null values in column home/away after the forward fill =',df['home/away'].isnull().sum())
```

```python
# If shot range is 16–24 ft → Shot type is Mid Range
mask = (df['range_of_shot'] == '16-24 ft.') & (df['shot_basics'].isna())
df.loc[mask, 'shot_basics'] = 'Mid Range'



#If shot range is Less Than 8 ft
# Randomly assign: Goal Area or Goal Line
# -------------------------------------------------------
mask = (df['range_of_shot'] == 'Less Than 8 ft.') & (df['shot_basics'].isna())

df.loc[mask, 'shot_basics'] = np.random.choice(
    ['Goal Area', 'Goal Line'],
    size=mask.sum(),
    p=[0.7590347263, 0.2409652737]
)



#If shot range is 8–16 ft
# Randomly assign: Mid Range or Goal Line
mask = (df['range_of_shot'] == '8-16 ft.') & (df['shot_basics'].isna())

df.loc[mask, 'shot_basics'] = np.random.choice(
    ['Mid Range', 'Goal Line'],
    size=mask.sum(),
    p=[0.6488754616, 0.3511245384]
)



#If shot range is 24+ ft
# Randomly assign: Penalty Spot, Right Corner, Left Corner
mask = (df['range_of_shot'] == '24+ ft.') & (df['shot_basics'].isna())

df.loc[mask, 'shot_basics'] = np.random.choice(
    ['Penalty Spot', 'Right Corner', 'Left Corner'],
    size=mask.sum(),
    p=[0.8932384342, 0.0619217082, 0.0448398577]
)



# If shot is Back Court Shot
# Randomly assign: Mid Ground Line or Penalty Spot
mask = (df['range_of_shot'] == 'Back Court Shot') & (df['shot_basics'].isna())

df.loc[mask, 'shot_basics'] = np.random.choice(
    ['Mid Ground Line', 'Penalty Spot'],
    size=mask.sum(),
    p=[0.8441558442, 0.1558441558]
)



#Check remaining missing values
df.isna().sum()
                                            
                                            
```

```python
df['shot_basics'].unique()
```

```python
# Filling Missing Values In "range_of_shot" based on "shot_basics" column

# if shot_basics is Goal Area, then range of shot is Less Than 8 ft
df.loc[df.shot_basics == 'Goal Area','range_of_shot']=(
    df.loc[df.shot_basics=='Goal Area','range_of_shot']
       .fillna('Less Than 8 ft.')
)

# if shot_basics is Penalty Spot, then range of shot is 24+ ft.
df.loc[df.shot_basics=='Penalty Spot','range_of_shot'] =(
    df.loc[df.shot_basics == 'Right Corner','range_of_shot']
       .fillna('24+ ft.')
)


```

```python
# Filling Missing Values In "range_of_shot" based on "shot_basics" column

# if shot_basics is Goal Area, then range of shot is Less Than 8 ft
df.loc[df.shot_basics == 'Goal Area', 'range_of_shot'] = (
    df.loc[df.shot_basics == 'Goal Area', 'range_of_shot']
      .fillna('Less Than 8 ft.')
)

# if shot_basics is Penalty Spot, then range of shot is 24+ ft.
df.loc[df.shot_basics == 'Penalty Spot', 'range_of_shot'] = (
    df.loc[df.shot_basics == 'Penalty Spot', 'range_of_shot']
      .fillna('24+ ft.')
)

# if shot_basics is Right Corner, then range of shot is 24+ ft.
df.loc[df.shot_basics == 'Right Corner', 'range_of_shot'] = (
    df.loc[df.shot_basics == 'Right Corner', 'range_of_shot']
      .fillna('24+ ft.')
)

# if shot_basics is Left Corner, then range of shot is 24+ ft.
df.loc[df.shot_basics == 'Left Corner', 'range_of_shot'] = (
    df.loc[df.shot_basics == 'Left Corner', 'range_of_shot']
      .fillna('24+ ft.')
)

# if shot_basics is Mid Ground Line, then range of shot is Back Court Shot
df.loc[df.shot_basics == 'Mid Ground Line', 'range_of_shot'] = (
    df.loc[df.shot_basics == 'Mid Ground Line', 'range_of_shot']
      .fillna('Back Court Shot')
)

# if shot_basics is Mid Range then randomly assign ranges
df.loc[
    (df.shot_basics == 'Mid Range') & (df.range_of_shot.isnull()),
    'range_of_shot'
] = np.random.choice(
    ['16-24 ft.', '8-16 ft.'],
    size=df[(df.shot_basics == 'Mid Range') & (df.range_of_shot.isnull())].shape[0],
    p=[0.6527708850289495, 0.34722911497105047]
)

# if shot_basics is Goal Line then randomly assign ranges
df.loc[
    (df.shot_basics == 'Goal Line') & (df.range_of_shot.isnull()),
    'range_of_shot'
] = np.random.choice(
    ['8-16 ft.', 'Less Than 8 ft.'],
    size=df[(df.shot_basics == 'Goal Line') & (df.range_of_shot.isnull())].shape[0],
    p=[0.5054360956752839, 0.49456390432471614]
)

# check missing values
df.isnull().sum()
```

## Making the Train and Test Dataset

```python
# train and test data are divided based on the vaue of is goal column
```

```python
# Making the train Dataset
train = df[df.is_goal.notnull()]
print('the Shape of Train Dataset',train.shape)
train.set_index(np.arange(train.shape[0]),inplace=True)
train.head()
```

```python
# Making the Test Dataset
test = df[df.is_goal.isnull()]
print('The Shape of Test Dataset',test.shape)
test.set_index(np.arange(test.shape[0]), inplace=True)
test.head()
```

## Handeling Missing Values in train and Test Datase

```python
#Filling the Nan value with a random choice from given list with there appropriate probablities
```

```python
l_goal = train[train.is_goal == 1].type_of_shot.value_counts().head(6).keys()
l_goal
```

```python
p_g_sum  = train[train.is_goal == 1].type_of_shot.value_counts().head(6).sum() # Top six shots when it was goal
p_goal   = (train[train.is_goal == 1].type_of_shot.value_counts().head(6) / p_g_sum ).tolist()  # There respective probablities
p_goal
```

```python
# if is_goal is 1, if type of shot is a string value, fill with the same or else fill with randomly choosing value from l_goal
g = pd.Series(train[train.is_goal == 1].type_of_shot.apply(lambda x: x if type(x)==str else np.random.choice(l_goal,1,p=p_goal)[0]))
g
```

```python
# # if is_goal is 1, if type of shot is null then type of shot becomes equal to the value of g based on the index
train.loc[(train.is_goal == 1)&(train.type_of_shot.isnull()), 'type_of_shot'] = g
```

```python
train['type_of_shot'].isna().sum()
```

```python
l_no_goal   = train[train.is_goal == 0].type_of_shot.value_counts().head(5).keys()     # Top five shots when it was not a goal
p_no_sum  = train[train.is_goal == 0].type_of_shot.value_counts().head(5).sum()
p_no_goal   = (train[train.is_goal == 0].type_of_shot.value_counts().head(5) / p_no_sum ).tolist() # There respective probablities 
ng = pd.Series(train[train.is_goal == 0].type_of_shot.apply(lambda x: x if type(x)==str else np.random.choice(l_no_goal,1,p=p_no_goal)[0]))
train.loc[(train.is_goal == 0)&(train.type_of_shot.isnull()), 'type_of_shot'] = ng 
train['type_of_shot'].isna().sum() # number of missing values got reduced to zero
```

```python
#Handeling the remaing values in test dataset with a smilira approach
test.loc[test['type_of_shot'].isnull(), 'type_of_shot'] = pd.Series(test.loc[test['type_of_shot'].isnull(), 'type_of_shot'].apply(lambda x: x if type(x)==str else np.random.choice(['shot - 39', 'shot - 36', 'shot - 4'],1,p=[0.37377133988618727, 0.33419555095706155, 0.2920331091567512])[0])) 
```
