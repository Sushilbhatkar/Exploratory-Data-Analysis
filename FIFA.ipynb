```python
#slecting columns to impute the missing values by mean 
to_impute_by_mean = df.loc[:, ['Crossing','Finishing', 'HeadingAccuracy', 'ShortPassing',
                               'Volleys', 'Dribbling','Curve', 'FKAccuracy', 'LongPassing',
                               'BallControl', 'Acceleration','SprintSpeed', 'Agility', 
                               'Reactions', 'Balance', 'ShotPower','Jumping', 'Stamina', 
                               'Strength', 'LongShots', 'Aggression','Interceptions', 
                               'Positioning', 'Vision', 'Penalties', 'Composure','Marking',
                               'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
                               'GKKicking', 'GKPositioning', 'GKReflexes']]
```

```python
#replacing the missing values with mean
for i in to_impute_by_mean.columns:
    df[i].fillna(df[i].mean(),inplace=True)
```

```python
##These are categorical variables and will be imputed by mode.
to_impute_by_mode = df.loc[:,['International Reputation','Height', 'Weight','Body Type',
                                'Preferred Foot','Jersey Number']]
```

```python
#replacing the missing values with mode
for i in to_impute_by_mode.columns:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
#The Following variables are either discreate numerical or continuous numerical variables.So the will be imputed by median.
to_impute_by_median = df.loc[:, ['Weak Foot', 'Skill Moves']]
```

```python
#replacing the missing values with median
for i in to_impute_by_median.columns:
    df[i].fillna(df[i].median(),inplace=True)
```

```python
#functions to get the rounded values from different columns
def defending(data):
    return int(round((data[['Marking','StandingTackle',
                            'SlidingTackle']].mean()).mean()))

def general(data):
    return int(round((data[['HeadingAccuracy','Dribbling','Curve',
                            'BallControl']].mean()).mean()))

def mental(data):
    return int(round((data[['Aggression','Interceptions','Positioning',
                            'Vision','Composure']].mean()).mean()))

def passing(data):
    return int(round((data[['Crossing','ShortPassing',
                            'LongPassing']].mean()).mean()))

def mobility(data):
    return int(round((data[['Acceleration','SprintSpeed',
                            'Agility','Reactions']].mean()).mean()))

def power(data):
    return int(round((data[['Balance','Jumping','Stamina',
                            'Strength']].mean()).mean()))

def rating(data):
    return int(round((data[['Potential','Overall']].mean()).mean()))

def shooting(data):
    return int(round((data[['Finishing','Volleys','FKAccuracy',
                            'ShotPower','LongShots','Penalties']].mean()).mean()))
```

```python
import requests
import random
from math import pi

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
def details(data, row, title, age, nationality, photo, image, logo, club):

    # temp image files
    flag_image = "img_flag.jpg"
    player_image = "img_player.jpg"
    logo_image = "img_club_logo.jpg"

    # download images
    requests.get(image, stream=True).raise_for_status()
    open(flag_image, 'wb').write(requests.get(image).content)

    open(player_image, 'wb').write(requests.get(photo).content)
    open(logo_image, 'wb').write(requests.get(logo).content)

    # colors
    r = lambda: random.randint(0, 255)
    colorRandom = '#%02X%02X%02X' % (r(), r(), r())
    if colorRandom == '#ffffff':
        colorRandom = '#a5d6a7'

    basic_color = '#37474f'
    annotate_color = '#01579b'

    # columns to ignore in radar
    ignore_cols = ['Name', 'Age', 'Nationality', 'Photo', 'Flag',
                   'Club_Logo', 'Club', 'ID']

    categories = [c for c in data.columns if c not in ignore_cols]
    values = data.loc[row, categories].values.tolist()

    N = len(categories)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]
    values += values[:1]

    # plot
    plt.figure(figsize=(15, 8))
    ax = plt.subplot(111, polar=True)
    ax.set_theta_offset(pi / 2)
    ax.set_theta_direction(-1)

    plt.xticks(angles[:-1], categories, size=14)
    ax.set_rlabel_position(0)
    plt.yticks([25, 50, 75, 100], ["25", "50", "75", "100"], color=basic_color)
    plt.ylim(0, 100)

    ax.plot(angles, values, color=basic_color, linewidth=2)
    ax.fill(angles, values, color=colorRandom, alpha=0.5)

    # background flag
    bg = mpimg.imread(flag_image)
    bg_ax = plt.gcf().add_axes([0, 0, 1, 1], zorder=-1)
    bg_ax.imshow(bg, alpha=0.3)
    bg_ax.axis('off')

    # annotations
    ax.annotate(f'Nationality: {nationality.upper()}',
                xy=(0, 0), xytext=(80, 180),
                textcoords='offset points',
                bbox=dict(facecolor=annotate_color, pad=6),
                color='white')

    ax.annotate(f'Age: {age}',
                xy=(0, 0), xytext=(80, 140),
                textcoords='offset points',
                bbox=dict(facecolor=annotate_color, pad=6),
                color='white')

    ax.annotate(f'Team: {club.upper()}',
                xy=(0, 0), xytext=(80, 100),
                textcoords='offset points',
                bbox=dict(facecolor=annotate_color, pad=6),
                color='white')

    # player image
    player_img = plt.imread(player_image)
    ab_player = AnnotationBbox(OffsetImage(player_img, zoom=0.5),
                               (0.5, 0.5), frameon=False)
    ax.add_artist(ab_player)

    # club logo
    logo_img = plt.imread(logo_image)
    ab_logo = AnnotationBbox(OffsetImage(logo_img, zoom=0.4),
                             (0.1, 0.1), frameon=False)
    ax.add_artist(ab_logo)

    plt.title(title, size=35, color=basic_color)
    plt.show()
```
